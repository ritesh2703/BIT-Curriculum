```python
df2=df.iloc[:2020,:]
df2
```

```python
df3=pd.DataFrame()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df3['Size'] = df2['Size']
df3['Weight']=df2['Weight']
df3['Sweetness']=df2['Sweetness']
df3['Softness'] = df2['Softness']
df3['HarvestTime'] = df2['HarvestTime']
df3['Ripeness'] = df2['Ripeness']
df3['Acidity'] = df2['Acidity']
df3['Quality']=encoder.fit_transform(df2['Quality'])
```

```python
df3
```

```python
X=df3.iloc[:,0:2]
X
```

```python
Y=df3.iloc[:,7:]
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
coef_Size, coef_Weight = model.coef_[0]
print(coef_Size, coef_Weight  )
```

```python
intercept = model.intercept_
intercept
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating a meshgrid for the plane
x_values = np.linspace(min(X["Size"]), max(X["Size"]), 10)
y_values = np.linspace(min(X["Weight"]), max(X["Weight"]), 10)
x_mesh, y_mesh = np.meshgrid(x_values, y_values)
```

```python
z_mesh = coef_Size * x_mesh + coef_Weight * y_mesh + intercept
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X["Size"], X["Weight"], Y)
# Plotting the best fit plane
ax.plot_surface(x_mesh, y_mesh, z_mesh, alpha = 0.3, color='green')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
```

```python
# Mean Square Error
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)
```

```python
# Mean Absolute Error
mae = mean_absolute_error(Y_test, Y_pred)
print("Mean Absolute Error:", mae)
```

```python
# R-squared score
r_squared = r2_score(Y_test, Y_pred)
print("R-squared score:", r_squared)
```
