# Experiment 10: Implement K-means Clustering algorithm and evaluate its performance 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
df.describe()
```

```python
X = df.drop(columns='species' , axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
means = KMeans(n_clusters=3)
km = KMeans(n_clusters = 3 ,random_state = 1)
km.fit(X)
```

```python
km.labels_
```

```python
km.cluster_centers_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X['sepal_length'], X['sepal_width'], c = km.labels_) 
plt.show() 
```

```python
plt.scatter(X['petal_length'], X['petal_width'], c = km.labels_) 
plt.show() 
```

```python
inertia_scores = [] 
for i in range(1,21): 
 k = KMeans (n_clusters = i) 
 k.fit(X) 
 inertia_scores.append(k.inertia_) 
```

```python
inertia_scores
```

```python
x_labels = range(1,21)
plt.plot(x_labels,inertia_scores,marker='o')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
cluster = AgglomerativeClustering(n_clusters=3).fit(X)
```

```python
cluster
```

```python
cluster.labels_
```

```python
plt.scatter(X['sepal_length'], X['sepal_width'], c = cluster.labels_) 
plt.show() 
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(km.labels_, cluster.labels_)
```
